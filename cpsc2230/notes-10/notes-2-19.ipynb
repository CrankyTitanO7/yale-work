# linked lists

- a collection of nodes that contain:
    - data ("payload")
    - a pointer to next node

we can define them as structs


```python
typedef struct list_node node;

struct list_node {
    int payload;
    node *next;
};

struct list_instance_t
{
    node *head; // the instance has a pointer to the head
};
```

note the use of head (not "first" or "first element"!)

so where does the last element node pointer point to?

... null. it becomes a null pointer.

Similarly, if a list is empty, the head value can be null, and there are no elements in the list.

see list create function:

```python
list *list_create {
    list *lst = malloc(sizeof(list));
    lst->head = NULL; // without this line, it is a garbage value.
    return lst;
}
```

## append
- linked list traversal
    - make a node, then link the last item pointer
    - efficiency O(n)

... unless you make the list backwards

- linked list append at the front
    - make a node whose pointer was head-pointer's value, then set head to new node
    - efficiency O(1) ALWAYS

## list destroy

you can:
### start at head
    - destroy each node on the way
    - end by freeing lst
### start at tail
    - using recusion:
        - you can destroy the nth node if the nth+1 node is destroyed...
        - and etc
    - end by destroying lst
