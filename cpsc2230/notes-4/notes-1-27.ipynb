## a short program that can read from the input and output streams

````python
#include <stdio.h>

int main()
{
    double x1, x2;
    fscanf(stdin, "%lf %lf", &x1, &x2);
    fprintf(stdout, "Average: %f\n", (x1 + x2) / 2);
    return 0;
}
````

• strcpy(dest, src) overwrites dest with characters from src
• strlen(s) counts the non-null characters in s
• strcat(dest, src) copies src to the end of dest
• strcmp(s1, s2) determines which of s1 and s2 comes first “ASCIIbetically”
• and more

so what is strcpy?

## strcpy

````python
void strcpy(char dest[], const char src[])
{
    size_t len = strlen(src);
    for (size_t i = 0; i < len; i++)
    {
        dest[i] = src[i];
    }
    dest[len] = '\0';
}
````

> const is a keyword that defines an immutable element

## strcat

````python
void strcat(char dest[], char src[])
{
    size_t i = 0;
    while (dest[i] != '\0')
        i++;
    size_t j = 0;
    while (src[j] != '\0')
        dest[i++] = src[j++];
    dest[i] = \0;

}
````
