### typedef

so sometimes we want to hide things in the main.c file

in list.h we can say

````python
// in list.h

typedef struct list_instance_t list;
````

and the program will assume it will happen. the implementation can be hidden in the implementation file:

````python
// in arraylist.c

#include "list.h"

struct list_instance_t{
    size_t length;
    size_t capacity;
    int *array;
};
````

note that things like 

````python
 else if (lst->length > 0) { // line 82
````

will break cause the main program can't access it! you must use a custom method in the implementation.c file

````python
// implementation.c
bool is_list_empty (const list *lst) {
    return lst->length == 0;
} 
````

````python
// main.c

 else if (is_list_empty (lst)) {
````

because implementation.c can "see" the struct in declaration.h BUT main.c cannot
