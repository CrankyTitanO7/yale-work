# pointers

- a type of variable that holds address information.

## semantic and syntax

- the & (ampersand) character is an operator that means address-of
- the type is noted *
- %p

````python
#include <stdio.h>
#include <stdlib.h>

int main (void) {
    int a = 10;
    int *ptr = &a;

    fprintf(stdout, "%p ", (void *) ptr); 
    // shows the address value of a
}

````

another method... 

````python
#include <stdio.h>
#include <stdlib.h>

int main (void) {
    int a = 10;
    int *ptr;

    ptr = &a;


    *(ptr) = 11;

    fprintf(stdout, "%zu ", (void *) ptr); 
    printf("%d\n", a);
    // shows the address value of a and changes its value to 11
}
````

what this does is 
- declare a value 'a'
- declare a value 'p', that holds the address of a

using *ptr makes it a reference

the * tells the compiler that this will be used as a pointer


fprintf is for specifying where to write out (better practice)

the (void \*) is an accessor. another use is 
```c 
printf("%zu", sizeof(int\*)) 
```
to see the pointer value precisely

## pointers for reference parameters

an example:

````python
int main() {
    int arr[] = {33, 20, 4, 42};
    int small, big;
    min_max(4, arr, &small, &big);
    printf("%d %d\n", small, big);
}
void min_max(size_t n, int *arr, int *min, int *max) {
    *min = arr[0];
    *max = *min;
    for (size_t i = 1; i < n; i++)
    if (arr[i] < *min)
    *min = arr[i];
    else if (arr[i] > *max)
    *max = arr[i];
}
````

# arrays and pointers 

A pointer can point to an array in memory
    • Actually, to the first item of the array
    • an array is a contiguous chunk of memory, indexable from zero (0)
• E.g., char *name = "Ozan"
    • name is a pointer to a char: 'O'
    • Part of a contiguous chunk of memory with enough space for 5 chars: 'O', 'z', 'a',
    'n', and '\0'
• Addressable with the subscript syntax ([])
    • name[0] is 'O' (equivalently *name)
    • name[1] is 'z'
    • name[2] is 'a', and so on…

````python
int main() {
    char hello[]= "Hello";      // declare as 'vanilla' array
    char *jello = "Jello";      // declare as pointer to first element of array

    char h = hello[0];          // ok
    hello[0] = 'h';             // ok
    hello = jello;              // NOT ok! there is no pointer labelled hello

    char j = jello[0];          // works, because I guess pointers are like arrays
    jello[0]= 'j';              // no, cannot assign. cannot alter a "string literal" (a type of immutable object). undefined behavior --> 
                                // likely segmentation fault.
    jello = hello;              // hello is not a pointer, but it can become one ("array decay into a pointer")
}
````
