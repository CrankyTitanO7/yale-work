## two methods of strcopy

````python
void strcpy(char dest[], const char src[])
{
    size_t len = strlen(src);
    
    for (size_t i = 0; i < len; i++)
    {
        dest[i] = src[i];
    }

    dest[len] = '\0';
}
````

````python
void strcpy(char *dest, const char *src)
{
while (*src != '\0')
{
    *dest = *src; // the second star dereferences the pointer to get the value at that address
    dest++;
    src++;
}

*dest = '\0';
}
````

## strlen pass by reference

````python
size_t strlen(const char *s)
{
    const char *p = s;

    while (*p != '\0')
    {
        p++;
    }
    
    return p - s;
}
````

p-s returns the length because it subtracts address of start from address of end

> note: above... p* (the values in the array) is constant, but p (the pointer) is not

# memory

• Memory is comprised of many bytes (8 bits)
• Each byte has an address
• We can use memory at consecutive addresses for larger things
• Your operating system organizes memory into blocks
• The Stack is one of those parts

the stack is typically memory with the highest addresses

• Each active function has a stack frame containing
    • arguments
    • local variables
    • some other things
• The calling function creates a stack frame on top of the stack
    • copies values into the arguments
• The called function adds its local variables  
• Stack frame is destroyed when the function returns
